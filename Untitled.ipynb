```python
from typing import List, Tuple

def solution(n: int, matches: List[Tuple[int, int, int, int]]) -> List[int]: # input i,j,p,q for each team of n; i and j (1 ≤ i < j ≤ n) are the IDs of the two teams, p is the number of goals scored by team i, q is the number of goals scored by team j.

    points = {i: 0 for i in range(1, n + 1)}
    goals_scored = {i: 0 for i in range(1, n + 1)}
    goals_conceded = {i: 0 for i in range(1, n + 1)} # create dictionaries where: keys = team IDs (1 to n) values = initial statistics (0)

    for i, j, p, q in matches: # logic to "evaluate" each match. i and j get assigned conceded and scored goals each.
        goals_scored[i] += p
        goals_scored[j] += q
        goals_conceded[i] += q
        goals_conceded[j] += p 

        if p > q: # i wins
            points[i] += 3
        elif p < q: # j wins
            points[j] += 3
        else: #equal goals/p=q
            points[i] += 1
            points[j] += 1


    teams = sorted(
    range(1, n + 1),   # teams from 1 to n

        
    key=lambda t: ( # ranking relys on negative num because sorted() sorts in accending order

        # higher points first
        -points[t],

        # goal difference
        -(goals_scored[t] - goals_conceded[t]),

        # scored goals
        -goals_scored[t],

        # else: smaller team id 
        t,
        ),
    )

    return teams
```

```python
from typing import List
def solution(inputList: List[int]) -> bool:
    inputLength = len(inputList)
    if inputLength == 1:
        return inputList[0] != 2 # if only one element, assume that a 2 inside cannot form a pair. if no two is included, treat as "buddy list"
    for i in range(inputLength):
        if inputList[i] == 2:
            previousNum = inputList[(i-1) % inputLength] # use modulo for circular behaviour of list
            nextNum = inputList [(i+1) % inputLength]
            if previousNum != 2 and nextNum != 2:
                return False
    return True # if no 2 is found or either previous or next number are a 2. 
```

```python
def solution (inputString: str) -> str:
    resultString = []
    inputLength = len(inputString)
    for i in range(inputLength):
        if i > 0:
            leftChar = inputString[i-1]
        else:
            leftChar = None
        if i < inputLength - 1:
            rightChar = inputString[i+1]
        else:
            rightChar = None
        if inputString[i] != leftChar and inputString[i] != rightChar:
            resultString.append(inputString[i])
    resultString = ''.join(resultString)
    return resultString
```

```python
import collections
def analyze_production_flow(edges):


    graph = collections.defaultdict(list) # store graph


    
    in_degree = collections.defaultdict(int)# how many prerequisites each task still has

    all_tasks = set() # store every node

    
    # -------------------------------------------------
    # STEP 2:
    # Build graph + compute in-degrees
    # -------------------------------------------------

    for before_task, after_task in edges:

        # remember both tasks exist
        all_tasks.add(before_task)
        all_tasks.add(after_task)

        # build directed edge
        #
        # before_task -> after_task
        graph[before_task].append(after_task)

        # after_task has one more prerequisite
        in_degree[after_task] += 1

        # ensure before_task exists in in_degree table
        #
        # tasks with no prerequisites
        # should still appear with value 0
        if before_task not in in_degree:
            in_degree[before_task] = 0



    # -------------------------------------------------
    # STEP 3:
    # Find all tasks that are immediately executable
    # -------------------------------------------------

    ready_tasks = []

    for task in all_tasks:

        # task has no prerequisites
        if in_degree[task] == 0:
            ready_tasks.append(task)



    # optional:
    # sort alphabetically for stable output
    ready_tasks.sort()



    # -------------------------------------------------
    # STEP 4:
    # Convert ready tasks into queue
    # -------------------------------------------------

    queue = collections.deque(ready_tasks)



    # -------------------------------------------------
    # STEP 5:
    # Store final execution order
    # -------------------------------------------------

    execution_order = []



    # -------------------------------------------------
    # STEP 6:
    # Process tasks one-by-one
    # -------------------------------------------------

    while queue:

        # take first ready task
        current_task = queue.popleft()

        # add to final schedule
        execution_order.append(current_task)



        # -------------------------------------------------
        # Every dependent task loses one prerequisite
        # -------------------------------------------------

        for dependent_task in graph[current_task]:

            # one prerequisite has been completed
            in_degree[dependent_task] -= 1



            # -------------------------------------------------
            # If no prerequisites remain:
            # task becomes executable
            # -------------------------------------------------

            if in_degree[dependent_task] == 0:

                queue.append(dependent_task)



    # -------------------------------------------------
    # STEP 7:
    # Detect cycle
    # -------------------------------------------------

    # If every task was processed:
    # graph is a valid DAG
    if len(execution_order) == len(all_tasks):

        return execution_order



    # otherwise:
    # some tasks were permanently blocked
    # -> cycle exists
    return "Logic Conflict: Deadlock Identified"



# -------------------------------------------------
# EXAMPLE
# -------------------------------------------------

edges = [
    ("A", "B"),
    ("B", "C"),
]

print(analyze_production_flow(edges))
```

```python
import math

def compute_triangle(a, b, c):
    if a >= b + c or b >= a + c or c >= a + b:
        return -1

    p = a + b + c
    s = p / 2

    area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    return (round(p, 2), round(area, 2))
```

```python
import numpy as np
def create_identity_matrix(n):
    ID = []
    for i in range (1,n+1):
        Row = []
        for k in range (1,n+1):
            if i == k:
                Row.append(float(1))
            else:
                Row.append(float(0))
        ID.append(Row)
    return np.array(ID)
```
