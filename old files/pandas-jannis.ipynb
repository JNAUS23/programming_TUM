```python


import numpy as np
import re
data = pd.read_csv('Topic3_healthcare_analytics_dataset.csv')
def to_midpoint(x):
    x = str(x).strip().lower()

    # range case: "10-20"
    m = re.match(r"(\d+)\s*-\s*(\d+)", x)
    if m:
        a, b = map(float, m.groups())
        return (a + b) / 2

    # open-ended case
    if "More than 100" in x:
        return 105  # or 105/110 depending on modeling choice

    # fallback
    return np.nan

data["Stay_num"] = data["Stay"].apply(to_midpoint)

data.groupby("Age")["Stay_num"].mean()
```
